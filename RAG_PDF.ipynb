```python
# Install necessary libraries
!pip install transformers langchain chromadb PyPDF2 sentence-transformers
!pip install -U langchain-community
!pip install pypdf

from transformers import AutoModelForCausalLM, AutoTokenizer
from langchain.chains import RetrievalQA
#Import PyPDFLoader from langchain.document_loaders
from langchain.document_loaders import PyPDFLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
from langchain.llms import HuggingFacePipeline
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
pdf_path = '/content/Project Catalyst - Milestone Module.pdf'
loader = PyPDFLoader(pdf_path)
documents = loader.load()
```

```python
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
vectorstore = Chroma.from_documents(documents, embeddings)
```

```python
model_name = "NousResearch/Llama-2-7b-chat-hf"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Create a pipeline with the model
pipe = pipeline("text-generation", model=model, tokenizer=tokenizer, max_new_tokens=100)
```

```python
!pip install -U langchain-huggingface
import langchain_huggingface
!pip install HuggingFacePipeline
```

```python
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
# Wrap the pipeline in a HuggingFacePipeline object for use with LangChain
llm = HuggingFacePipeline(pipeline=pipe)
```

```python
qa_chain = RetrievalQA.from_chain_type(llm=llm, chain_type="stuff", retriever=vectorstore.as_retriever())
```

```python
query = "What is the main topic of the PDF?"
response = qa_chain.run(query)
print(response)
```
